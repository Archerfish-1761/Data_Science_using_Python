## Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([3,13,19,31,38,45,58,65,71,88])
y = np.array([11,16,24,37,41,58,56,78,81,95])
z = np.array([81,80,79,63,54,32,29,26,18,14])
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.plot(x,y)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.plot(x,y,z)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x,y, color='red', marker='x', ls='-', mfc='green')   #mfc = marker face color, ls= list
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x,y, color='red', marker='x', label='2022', mfc='green')
plt.plot(x,z, color='orange', marker='x', label='2023', mfc='green')
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
```

```python
plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(x,z)
```

```python
plt.figure(figsize=(24,10))
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x,y, lw =5, marker= 'o',ms=20, mfc='y')
plt.savefig('myplot.png')  #can only be saved in png format
```

## Bar Plot

```python
enrollment = [104,112,256,178,126,180]
courses = ['C','Java','Python','R','PHP','JavaScript']
```

```python
plt.figure(figsize=(16,10))
plt.xlabel('Courses')
plt.ylabel('Enrollment')
plt.grid()
plt.bar(courses,enrollment,color='Green')
```

```python
plt.figure(figsize=(16,10))
plt.ylabel('Courses')
plt.xlabel('Enrollment')
plt.grid()
plt.barh(courses,enrollment,color='Orange')
```

## Pie Chart

```python
plt.figure(figsize=(16,10))
plt.pie(enrollment,labels=courses, autopct='%2.2f');  #autopct= autopercentage
```

## Univariate Analysis

### Histogram

```python
marks = np.array([67,56,45,61,93,72,65,58,65,49,68,62])
```

```python
plt.figure(figsize=(16,10))
plt.title('Frequency plot of Marks')
plt.xlabel('Range of Marks')
plt.ylabel('Count')
plt.grid()
plt.xticks(range(0,101,10))
plt.hist(marks, bins= range(0,101,10),color='pink')
```

## Box Plot

```python
plt.boxplot(marks);
```
